```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
raw_data = pd.read_csv('Glacier_Peak_Data.csv')

def remove_percentage(s):
    if isinstance(s, str):
        return float(s.rstrip('%'))
    else:
        return 0

raw_data['How much defense played?'] = raw_data['How much defense played?'].apply(remove_percentage)
```

```python
data = raw_data.groupby('Team #')['# cubes on scale:', '# cubes on our switch:', '# cubes on opponent’s switch:',
                                  '# cubes into exchange:', 'How much defense played?'].mean()

data['total cubes'] = data['# cubes on scale:'] + data['# cubes on our switch:'] \
    + data['# cubes on opponent’s switch:'] + data['# cubes into exchange:']

data = data.sort_values(by='total cubes', ascending=False)
```

```python
plot1 = data[['# cubes on scale:', '# cubes on our switch:', '# cubes on opponent’s switch:',
              '# cubes into exchange:']].plot(kind='bar', stacked=True, title='Tele Cubes')

plot1.set_xlabel('Team')
plot1.set_ylabel('Total cubes');
```

```python
plot2 = data['# cubes on scale:'].plot(kind='bar', stacked=True, title='Scale Cubes', color='green')
plot2.set_xlabel('Team')
plot2.set_ylabel('Avg Scale');
```

```python
plot3 = data['# cubes on our switch:'].plot(kind='bar', stacked=True, title='Own Switch Cubes', color='blue')
plot3.set_xlabel('Team')
plot3.set_ylabel('Avg Own Switch');
```

```python
plot4 = data['# cubes on opponent’s switch:'].plot(kind='bar', stacked=True, title='Opp Switch Cubes', color='red')
plot4.set_xlabel('Team')
plot4.set_ylabel('Avg Opp Switch');
```

```python
plot5 = data['# cubes into exchange:'].plot(kind='bar', stacked=True, title='Exchange Cubes', color='yellow')
plot5.set_xlabel('Team')
plot5.set_ylabel('Avg Exchange');
```

```python
plot6 = data['How much defense played?'].plot(kind='bar', stacked=True, title='Defense', color='magenta')
plot6.set_xlabel('Team')
plot6.set_ylabel('Avg Defense');
```
